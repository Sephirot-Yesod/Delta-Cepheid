```python
import urllib.request
import certifi
import ssl

def parse_int(line: str, index: int) -> float:
    f, dec, idx, sign = 0, 0, 0, 1

    for x in range(index, len(line)):
        if (line[x] == ' ' or line[x] == '\r'):
            break
        elif (line[x] == '.'):
            dec = (x - index)
        elif (line[x] == '-'):
            sign = -1
        else:
            f = 10 * f + int(line[x])
        idx += 1

    return sign * (f / (10 ** (idx - dec - 1)))

def req():
    url = 'https://www.astro.utoronto.ca/DDO/research/cepheids/table_physical.html'
    resp = urllib.request.urlopen(url, context=ssl.create_default_context(cafile=certifi.where()))
    data = resp.read()
    contents = data.decode('utf-8').strip().split('\n')[13:522]

    dat = { "period": [], "MV": [] }

    # period starts from 15 
    # MV starts from 59
    for x in range(len(contents)):
        dat["period"].append(parse_int(contents[x], 15))
        dat["MV"].append(parse_int(contents[x], 59))

    return dat

if (__name__  == "__main__"):
    dat = req()

    print(dat['period'])
    print(dat['MV'])
```
